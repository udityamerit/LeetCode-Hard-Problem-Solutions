```python
class Solution:
    def countSubarrays(self, nums: List[int], k: int) -> int:
        ##Time limit exceeded at test case no 157
        # n = len(nums)
        # l = 0
        # for a in range(1, n+1):
        #     for i in range(n-a+1):
        #         subarr = nums[i:i+a]
        #         s = sum(subarr)
        #         m = len(subarr)
        #         if s*m<k:
        #             l = l+1
        # # print(l)
        # return 
        
        total = 0
        window_sum = 0
        left = 0
        for right, num in enumerate(nums):
            window_sum += num
            while window_sum * (right - left + 1) >= k:
                window_sum -= nums[left]
                left += 1
            total += (right - left + 1)
        return total
```
